# Q10

10. How can you tell the difference between break and continue?

Ans. Break --> take the program out of the loop with out any further iteration
    continue --> skips only the particular iteration on which it was hit and executes the remaining loop as it is
    

```python
#Example program for break and continue
num = 0

while(num<10):
    
    num+=1
    if num==5:
        continue #Skip the step when num == 5
    if num == 8:
        break # come out of the loop when num == 8
    print(num)
```
